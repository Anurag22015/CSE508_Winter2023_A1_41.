```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
  
inverted_index={}
def bigram_index(file_path):
  f = open(file_path, 'r')
  word=f.read()
  tokens=word_tokenize(word)
  l=len(tokens)
  st=file_path.split("cranfield",1)[1]
  for i in range (0,l-1):
    key=tokens[i]+' '+tokens[i+1]
    if key not in inverted_index:
      inverted_index[key] = list()
      inverted_index[key].append(re.findall(r'\d+', st))
    else:
      t=re.findall(r'\d+', st)
      if inverted_index[key][-1]!=t:
        inverted_index[key].append(re.findall(r'\d+', st))
  f.close()
def find_freq_dict():
  for key in inverted_index:
    t=len(inverted_index[key])
    inverted_index[key].insert(0,t)
for i in range (1,1401):
       bigram_index('/content/drive/MyDrive/CSE508_Winter2023_Dataset/cranfield'+str('{0:04}'.format(i)))
find_freq_dict()
```

```python
def intersection_pos(list1,list2):
  intersection=list()
  freq1=list1[0]
  freq2=list2[0]
  freq=0
  for i in range (1,freq1+1):
    item=list1[i][0]
    for j in range (1,freq2+1):
      if item==list2[j][0]:
        l=len(list1[1])
        for k in range (1,l):
          item2=list1[i][k]+1
          if item2 in list2[i]:
            intersection.append(list())
            intersection[-1].append(item)
            intersection[-1].append(item2)
            freq+=1
  intersection.insert(0,freq)
  return intersection

def intersection_bi(list1,list2):
  intersect=list()
  freq1=list1[0]
  freq2=list2[0]
  freq=0
  for i in range (1,freq1+1):
    item=list1[i]
    for j in range (1,freq2+1):
      if item==list2[j]:
          intersect.append(item)
          freq+=1
  intersect.insert(0,freq)
  return intersect


```
